## 3. Organize All The Data In a Dictionary

```python
 # Create a dictionary with original name of league as key
dict_countries ={ 'Spanish La Liga' :'SP1' , 'Spanish Segunda Division': 'SP2',
                  'German Bundesliga' : 'D1',
                   'English Premier League' : 'E0','English League 1' :'E2','English League' :'E3'}
```

```python
# get  a dictionary element
dict_countries['Spanish La Liga']
```

```python
# looping through the dictionary
for dic in dict_countries:
    print(dic)
```

```python
for dic in dict_countries:
    print(dict_countries[dic])
```

```python
dict_historical_data = {}


#looping through key element
for league in dict_countries:
    frames = []
    for season in range(20,26):
        
        df= pd.read_csv( root + str(season) + str(season+1) + '/' + dict_countries[league] + '.csv')
        df.insert(1,'Seasons', season)
        frames.append(df)
    df_concat =pd.concat(frames)  
    dict_historical_data[league] = df_concat 


        
```

 # 1. Creating a conditional column from More than 2 choices : np.select()

```python
df_laptops.head(3)
```

```python
# create an array based on multiple price tiers ( +2choices)

#create "conditions' and "values" variables

conditions = [
df_laptops['Price_euros']>3000,
(df_laptops['Price_euros']>2000) & (df_laptops['Price_euros']<=3000),
(df_laptops['Price_euros']>800) & (df_laptops['Price_euros']<=2000),
df_laptops['Price_euros']<=800 
]
values = ['Too Expensive','Expensive','Affordable','Cheap']
```

```python
# set it to a new column
df_laptops['Price_Tiers'] = np.select(conditions,values, default = 'Unknown')
```

```python
# show dataframe
df_laptops
```

```python
df_laptops['Price_Tiers'].value_counts()
```
